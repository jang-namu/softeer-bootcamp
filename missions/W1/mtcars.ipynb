# gear와 transmission의 경우의 수 조합별로 나누고 각 조합에 몇개의 차량이 있는지를 알아내세요.

먼저 결과값을 화면에 출력하세요.

결과값을 바 그래프로 그리세요. 타이틀은 '# of Cars by Gear x Transmission'로, x축 레이블은 '(Gear, Auto Transmission)', y축 레이블은 '# of Cars'로 만드세요.

```python
fc2
```

```python
dup_gear_transmissions = fc2[['gear', 'am']]
```

```python
dup_gear_transmissions
```

```python
gear_transmissions_group=dup_gear_transmissions.groupby(by=['gear', 'am']).size().reset_index().rename(columns={0:'count'})
```

```python
gear_transmissions_group
```

```python
df = gear_transmissions_group
```

```python
xs = df[['gear', 'am']].values
```

```python
xs = ["(" + str(x) + ", " + str(y) +")" for x, y in xs]
```

```python
xs
```

```python
plt.bar(xs, df['count'])
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')
plt.title('# of Cars by Gear x Transmission')
plt.show()
plt.close()
```

# 데이터셋에 포함된 모든 features의 히스토그램을 그리세요.

그래프 타이틀은 "Histograms of Variables"로 하세요.

최대한 잘 보이도록 그래프들을 배치하세요.

```python
fc2
```

```python
fc3=fc2.drop('cars_name', axis=1)


fig, axes = plt.subplots(3, 4, figsize=(16, 16))
fig.suptitle("Histograms of Variables", fontsize=24)

i = 1
for col in fc3:
    plt.subplot(3, 4, i)
    plt.hist(fc3[col])
    plt.title(col)
    plt.subplots_adjust(bottom=0.3, wspace=0.2, hspace=0.2)
    i += 1
plt.show()
```
